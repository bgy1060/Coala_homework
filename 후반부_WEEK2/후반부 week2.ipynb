#### 6). 판다스로 csv 데이터 불러온 후 seaborn 세팅하기

```python
import pandas as pd

train=pd.read_csv('data/train.csv')

train
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

#### 7). 기본적인 산포도 그리기

```python
sns.relplot(data=train,x='Age',y='Survived')
```

#### 8). 범주 데이터 넣기 & 사이즈 조절

```python
sns.relplot(data=train,x='Age',y='Survived',hue='Sex',aspect=2)

sns.relplot(data=train,x='Age',y='Survived',hue='Sex',aspect=5)
```

#### 9). Categorical Scatter Plot 그려보기

```python
sns.catplot(data=train,x='Pclass',y='Fare',aspect=4)

sns.catplot(data=train,x='Pclass',y='Fare',jitter=False,aspect=4)

sns.catplot(data=train,x='Pclass',y='Fare',jitter=True,aspect=4)
# jitter=True를 사용하면 데이터 수가 많을 경우 겹치지 x

sns.catplot(data=train,x='Pclass',y='Fare',aspect=4,hue='Sex')
```

#### 10). Pair plot 기능 사용하기

```python
sns.pairplot(data=train,hue='Sex',
             x_vars=['Pclass','Age','Fare'],
             y_vars=['Survived','Age','Fare'],height=3)
```

#### 11). Box Plot 그리기

```python
sns.catplot(data=train,x='Pclass',y='Fare',kind='box')
sns.catplot(data=train,x='Pclass',y='Fare',kind='box',hue='Sex')
```
